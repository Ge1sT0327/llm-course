# 4.1 Agent设计模式

实现ReActAgent和PlanSolveAgent - 智能知识库Agent的推理大脑

## 项目定位
ReActAgent和PlanSolveAgent被6.3最终项目的KnowledgeBaseAgent直接使用

```python
import sys, os
sys.path.insert(0, "..")
from config import get_client; client = get_client()
from agent_project import *
from agent_project.tools import create_default_registry
print(f"LLM: {client.name} | project modules loaded")
```

```python
from agent_project.agent import ReActAgent, PlanSolveAgent
tools = create_default_registry()
react = ReActAgent(client, tools, max_steps=5, verbose=True)
plan = PlanSolveAgent(client, tools, verbose=True)
print(f"Agents ready: ReAct + PlanSolve, tools={tools.list_tools()}")
```

```python
print("===== ReAct Agent: Multi-step Reasoning =====\n")
answer = react.run("Search for MCP protocol definition, then calculate sqrt(256)")
print(f"\nFinal Answer: {answer[:300]}")
print(f"Steps: {len(react.trace)}")
for s in react.trace:
    if s.action:
        print(f"  Step{s.step_num}: {s.action} -> {s.observation[:60]}")
```

```python
print("\n===== PlanSolve Agent: Plan then Execute =====\n")
result = plan.run("Compare DeepSeek V4 and Qwen3.7 for code generation")
print(f"Plan: {len(result["steps"])} steps")
for r in result["results"]:
    print(f"  Step{r["step"]}: {r["result"][:100]}...")
```
